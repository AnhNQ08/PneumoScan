# Exploratory Data Analysis (EDA)

Notebook này dùng để khám phá dữ liệu trước khi train model. Việc này giúp ta hiểu rõ dữ liệu, phát hiện lỗi (nếu có) và lựa chọn chiến lược augmentation phù hợp.

```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import random

# Cấu hình đường dẫn
data_dir = '../data/raw/train'
classes = ['NORMAL', 'PNEUMONIA']

def show_samples(data_dir, num_samples=5):
    plt.figure(figsize=(15, 5))
    for i, class_name in enumerate(classes):
        class_dir = os.path.join(data_dir, class_name)
        if not os.path.exists(class_dir):
            print(f"Directory {class_dir} not found. Please download data first.")
            continue
            
        image_files = os.listdir(class_dir)
        samples = random.sample(image_files, min(len(image_files), num_samples))
        
        for j, img_name in enumerate(samples):
            img_path = os.path.join(class_dir, img_name)
            img = Image.open(img_path)
            
            plt.subplot(2, num_samples, i * num_samples + j + 1)
            plt.imshow(img, cmap='gray')
            plt.title(class_name)
            plt.axis('off')
    plt.show()

show_samples(data_dir)
```

```python
# Kiểm tra số lượng ảnh mỗi lớp
for class_name in classes:
    path = os.path.join(data_dir, class_name)
    if os.path.exists(path):
        print(f"{class_name}: {len(os.listdir(path))} images")
    else:
        print(f"{class_name}: 0 images (Not found)")
```
